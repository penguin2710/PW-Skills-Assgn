In Bokeh, glyphs refer to the visual markers or shapes used to represent data points on a plot. They can be basic shapes such as circles, squares, or triangles, as well as more complex shapes like polygons or even custom glyphs.

To add glyphs to a Bokeh plot, you can use the figure() function to create a plot, and then call the appropriate glyph method, such as circle(), square(), or triangle(), to specify the glyph type and customize its properties.

```python
#Ans 2

from bokeh.sampledata.iris import flowers

p = figure()

p.circle(flowers['sepal_length'], flowers['sepal_width'], size=8, fill_color='red')

p.square(flowers['petal_length'], flowers['petal_width'], size=8, fill_color='blue')

p.triangle(flowers['sepal_length'], flowers['petal_width'], size=8, fill_color='green')

show(p)
```

```python

#Ans 3
p = figure(plot_width=500, plot_height=400, title="Iris Dataset", x_axis_label='Petal Length',
           y_axis_label='Petal Width')

p.circle(flowers[flowers['species'] == 'setosa']['petal_length'],
         flowers[flowers['species'] == 'setosa']['petal_width'], size=8, color='red', alpha=0.5, legend_label='Setosa')

p.circle(flowers[flowers['species'] == 'versicolor']['petal_length'],
         flowers[flowers['species'] == 'versicolor']['petal_width'], size=8, color='green', alpha=0.5, legend_label='Versicolor')

p.circle(flowers[flowers['species'] == 'virginica']['petal_length'],
         flowers[flowers['species'] == 'virginica']['petal_width'], size=8, color='blue', alpha=0.5, legend_label='Virginica')

# Customize the appearance
p.title.align = 'center'
p.title.text_font_size = '18px'
p.title.text_color = 'navy'

p.xaxis.axis_label_text_font_size = '14px'
p.yaxis.axis_label_text_font_size = '14px'

p.legend.location = 'top_left'
p.legend.label_text_font_size = '12px'

# Display the plot
show(p)
```
