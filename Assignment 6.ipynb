```python
#Ans 1

'''A class is a user-defined blueprint that defines a set of attributes (variables) and functions
(methods) that an object of that class will possess.'''
'''An object is an instance of a class. It represents a specific entity created from the class blueprint.'''

class information:          #creating a class
    name="Pratham"        
    age=21
    occ="Job"
    
    def info(self):       #a method
        print(f"{self.name} is a {self.occ}")

a = information()                #creating an object
print(a.name, a.age, a.occ)

b = information()           #creating new object for same class
b.info()
```

```python
#Ans 5

"Inheritance is being able to use the properties of parent class in child class"

#Single Inheritance
class Employee:
  def __init__(self, name, id):
    self.name = name
    self.id = id 

  def showDetails(self):
    print(f"The name of Employee: {self.id} is {self.name}")

class Programmer(Employee):                  #this new class will have and run all attributes of employee
  def showLanguage(self):
    print("The default langauge is Python")


#Multiple Inheritance
class Mother:
    mothername = ""
 
    def mother(self):
        print(self.mothername)
 
class Father:
    fathername = ""
 
    def father(self):
        print(self.fathername)
 
class Son(Mother, Father):
    def parents(self):
        print("Father name is :", self.fathername)
        print("Mother :", self.mothername)


#Hybrid Inheritance
class School:
    def func1(self):
        print("This function is in school.")
 
class Student1(School):
    def func2(self):
        print("This function is in student 1. ")
 
class Student2(School):
    def func3(self):
        print("This function is in student 2.")
 
class Student3(Student1, School):
    def func4(self):
        print("This function is in student 3.")


#Multilevel Inheritance
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def show_details(self):
        print(f"Name: {self.name}")
        print(f"Species: {self.species}")
        
class Dog(Animal):
    def __init__(self, name, breed):
        Animal.__init__(self, name, species="Dog")
        self.breed = breed

    def show_details(self):
        Animal.show_details(self)
        print(f"Breed: {self.breed}")
        
class GoldenRetriever(Dog):
    def __init__(self, name, color):
        Dog.__init__(self, name, breed="Golden Retriever")
        self.color = color

    def show_details(self):
        Dog.show_details(self)
        print(f"Color: {self.color}")


#Heirarchichal
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Animal speaks.")


class Dog(Animal):
    def speak(self):
        print("Dog barks.")


class Cat(Animal):
    def speak(self):
        print("Cat meows.")
```
