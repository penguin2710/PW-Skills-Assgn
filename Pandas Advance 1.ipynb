```python
#Ans 10

data2 = {'Date': ['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05', '2023-07-06', '2023-07-07'],
        'Sales': [100.0, 75.0, 120.0, 80.0, 110.0, 150.0, 90.0]}
df4 = pd.DataFrame(data2)

df4['Date']= pd.to_datetime(df4['Date'])
df4.sort_values('Date', inplace=True)

def movavg():
    df4['Moving Average'] = df4['Sales'].rolling(window=7, min_periods=1).mean()
    return df4

movavg()
```

```python
#Ans 11

data3 = {'Date':['2023-07-01', '2023-07-02', '2023-07-03', '2023-07-04', '2023-07-05', '2023-07-06', '2023-07-07']}

df5 = pd.DataFrame(data3)
df5['Date']= pd.to_datetime(df5['Date'])

def weekday():
    df5['Weekday'] = df5['Date'].dt.strftime('%A')
    return df5

weekday()
```

```python
start_date = pd.to_datetime('2022-12-01')
end_date = pd.to_datetime('2023-02-28')

dates = pd.date_range(start_date, end_date)

df6 = pd.DataFrame({'Dates':dates})
print(df6)

def date_sort():
    df7 = df6[(df6['Dates']>='2023-01-01') & (df6['Dates']<='2023-01-31')]
    return df7

date_sort()
```
