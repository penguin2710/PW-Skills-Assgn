Matplotlib is a library in python which is used to plot different types of graphs and charts for data visualization. It is used to interpret the data more clearly. Types of graphs that can be plotted:

1. Line Chart
2. Scatter Plot
3. Bar Chart
4. Pie Chart
5. Histogram

```python
#Ans 2
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.ylabel("y axis")
plt.title("Scatter plot")
plt.xlabel("x axis")
plt.scatter(x,y)
```

```python
#Ans 3
x1 = np.array([0, 1, 2, 3, 4, 5]); y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5]); y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5]); y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5]); y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('Line 1')

plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.title('Line 2')

plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('Line 3')

plt.subplot(2,2,4)
plt.plot(x4,y4)
plt.title('Line 4')
```

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It displays the five-number summary of the data, which includes the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum.

A box plot is used to provide a concise summary of the distribution, allowing for easy visual comparison between multiple datasets. It can reveal information about the skewness, outliers, and the overall spread of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]
plt.boxplot(data)
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot')
plt.show()
```
