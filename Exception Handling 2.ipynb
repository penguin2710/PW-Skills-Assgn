using the Exception class as the base for your custom exception helps ensure consistency, compatibility, and maintainability in your code, making it easier to handle and reason about exceptions in a standardized manner.

```python
#Ans 2

def print_exception_hierarchy(exception_class, indent=0):
    print(' ' * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(Exception)
```

```python
#Ans 3

#1. ZeroDivisionError
try:
    result = 10 / 0  # Division by zero
    print(result)
except ZeroDivisionError:
    print("Error: Division by zero!")

#2. InvalidOperation
from decimal import Decimal, InvalidOperation

try:
    result = Decimal('abc') + Decimal('1.23')
    print(result)
except InvalidOperation:
    print("Error: Invalid arithmetic operation!")
```

```python
#Ans 4

#KeyError
my_dict = {'name': 'John', 'age': 30}

try:
    value = my_dict['address'] 
except KeyError as e:
    print(e)

#IndexError
my_list = [10, 20, 30]

try:
    value = my_list[3] 
except IndexError as e:
    print(e)
```
